```python
import os, sys, pickle
import nltk
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('testing_data.csv')
df
```

```python
df['question'] = df['question'].apply(nltk.word_tokenize)
df
```

```python
with open('valid_words.pickle', 'rb') as f:
    valid_words = pickle.load(f)
```

```python
def f(arr):
    arr = [w.lower() for w in arr if w.isalnum() or w == '_____']
    for i in range(len(arr)):
        if arr[i].startswith('__'):
            targetIndex = i
            break
    res = arr[max(0, targetIndex-10):targetIndex]
    return res

df['testingData'] = df['question'].apply(f)
df['testingData']
```

```python
def f2(arr):
    return list(map(lambda w: w if w in valid_words else 'UNK', arr))
df['testingData'] = df['testingData'].apply(f2)
df['testingData']
```

```python
ans = pd.read_csv('test_answer.csv')
df = df.merge(ans, on='id', how='left')
df
```

```python
output = df[['testingData', 'a)', 'b)', 'c)', 'd)', 'e)', 'answer']]
output.to_csv('parsed_testing_data.csv', sep='\t', encoding='utf-8', index=False)
```

```python
testingX = output['testingData'].tolist()
with open('testingX.pkl', 'wb') as f:
    pickle.dump(testingX, f)
```
